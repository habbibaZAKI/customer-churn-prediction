```python
# Collect user input
user_input = {}

user_input['tenure'] = int(input("Enter tenure (months): "))
user_input['MonthlyCharges'] = float(input("Enter monthly charges: "))
user_input['TotalCharges'] = float(input("Enter total charges: "))

user_input['Contract'] = input("Contract type (Month-to-month / One year / Two year): ")
user_input['PaymentMethod'] = input("Payment method (Electronic check / Mailed check / Bank transfer / Credit card): ")
user_input['InternetService'] = input("Internet service (DSL / Fiber optic / No): ")

# Binary features
user_input['gender'] = input("Gender (Male / Female): ")
user_input['Partner'] = input("Partner? (Yes / No): ")
user_input['Dependents'] = input("Dependents? (Yes / No): ")
user_input['PhoneService'] = input("PhoneService? (Yes / No): ")
user_input['PaperlessBilling'] = input("PaperlessBilling? (Yes / No): ")
```

```python
binary_map = {'Yes': 1, 'No': 0, 'Female': 1, 'Male': 0}

for col in ['gender','Partner','Dependents','PhoneService','PaperlessBilling']:
    user_input[col] = binary_map[user_input[col]]
```

```python
categorical_cols = ['PaymentMethod','Contract','InternetService']

# Convert to DataFrame
input_df = pd.DataFrame([user_input])

# One-hot encode
input_encoded = pd.get_dummies(input_df, columns=categorical_cols, drop_first=True)
```

```python
import pickle

# Load the feature columns
with open("feature_columns.pkl", "rb") as f:
    feature_columns = pickle.load(f)

# Reindex the user input DataFrame
input_encoded = input_encoded.reindex(columns=feature_columns, fill_value=0)
```

```python
input_encoded = input_encoded.reindex(columns=feature_columns, fill_value=0)
```

```python
import pickle

with open("churn_model.pkl", "rb") as f:
    model = pickle.load(f)

print("Model loaded successfully!")
```

```python
# Predict probability
prob = model.predict_proba(input_encoded)[0][1]  # probability of Churn = 1

# Predict class
prediction = model.predict(input_encoded)[0]

# Output results
if prediction == 1:
    print(f"Customer is likely to CHURN! Probability: {prob:.2f}")
else:
    print(f"Customer is likely to STAY! Probability: {prob:.2f}")
```
